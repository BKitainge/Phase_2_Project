2. *Average Price Vs View*

*  The average price of a house with no view is 540,000 dollars
*  The average price of a house with an average view is 1,100,000 dollars
* The average price of a house with a good view is 1,500,000 dollars
* The average price of a house with a fair view is 1,200,000 dollars
* The average price of a house with an excellent view is 2,000,000 dollars
* The average price of a house with a waterfront is 1,800,000 dollars
* The average price of a house with no waterfront is 540,000 dollars
* The price increases as the view increases, however it seems houses with a good view ```2``` are priced higher than houses with a fair view ```3```
